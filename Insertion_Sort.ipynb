**Insertion Sort**

In Bubble sort:-  
-> we use two loops (one inner and one outer).  
-> the outer loop (say i)is iterated from the 1st index to the last index.  
-> we will take a **key** varable and store the value of **arr[i]**.  
-> we will now take a varable **j** which will point towards the **0th** index.   
-> we will use while loop as inner loop and compare the value of j ( j >= 0 and arr[j] > key).  
-> The iterations will be repeated until the value of i reaches the last index.

````python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        # after this statement, j will not point to 0th index after 1st iteration
        # j will point to -1 index
        arr[j + 1] = key
        # after this j will again point to 0th index
        # and value of the key will be substituted in the 0th index

    return arr
arr = [13, 46, 24, 52, 20, 9]
sorted_arr = insertion_sort(arr)
print(sorted_arr)
````
