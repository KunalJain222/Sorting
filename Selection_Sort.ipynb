**Selection Sort**

In selection sort:-  
-> we use two loops (one inner and one outer).  
->the outer loop takes the starting point as minimum no.  
->the inner loop is iterated with the rest of the array. If the value of the element inside the right side of the array is less than the minimum , then switch .  
-> increment the array from the beginning and the end.  
-> print the sorted array.

Code 

````python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):

        mini_index = i

        for j in range(i+1,n):
            if arr[j] < arr[mini_index]:
                mini_index = j

        #temp = arr[mini_index]
        #arr[mini_index] = arr[i]
        #arr[i] = temp
        #instead of using temp we can directly swap the elements as well
        arr[i], arr[mini_index] = arr[mini_index], arr[i]
    return arr

arr = [64, 25, 12, 22, 11]
sorted_arr = selection_sort(arr)
print("sorted array:", sorted_arr)
````

Taking input from the user 

````python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

# Taking input from the user
user_input = input("Enter numbers separated by spaces: ")
arr = list(map(int, user_input.split()))

sorted_arr = selection_sort(arr)
print("Sorted array:", sorted_arr)
````

Q. What is the use of :-

```
arr = list(map(int, user_input.split()))
```
Ans. This line is taking a string of space-separated numbers from the user and converting it into a list of integers.  
Here's how it works step-by-step:


1.   user_input.split()  
Suppose the user enters:  
```
29 10 14 37 13
```
split() without any arguments splits the string by spaces by default.  
So this gives:
```
['29', '10', '14', '37', '13']
```
This is a list of strings.

2.   map(int, user_input.split()):  
map() takes a function and applies it to every element in an iterable.  
Here, int is the function, and user_input.split() gives the iterable (a list of string numbers).  
So, map(int, ['29', '10', '14', '37', '13']) turns each string into an integer.  
You get something like:  
```
map object  [29, 10, 14, 37, 13]
```

3.   list(...)  
list() is used to convert the map object into a proper Python list.  
Final result:
```
arr = [29, 10, 14, 37, 13]
```
