**Merge Sort**

In Merge Sort we use:-  
-> **divide and conquers algorithm**.  
-> It divides the given array into equal parts and then merges the 2 sorted parts.  
-> There are 2 main functions :  
**merge()**: This function is used to merge the 2 halves of the array. It assumes that both parts of the array are sorted and merges both of them.  
**mergeSort()**: This function divides the array into 2 parts. low to mid and mid+1 to high where,  
 low = leftmost index of the array  
 high = rightmost index of the array  
 mid = Middle index of the array   
-> We recursively split the array, and go from top-down until all sub-arrays size becomes 1.

Here's step-by-step approach:-
1. Merge Sort Approach – Step-by-Step  

Step 1: Create Two Functions  


```
mergeSort(arr, low, high)
```



```
merge(arr, low, mid, high)  
```
Step 2:
```
 mergeSort(arr, low, high)
```
1.  **Base Case:**

*   If
```
low >= high
```
 , return (only one element).

2.  **Recursive Case:**

*   Find
```
mid = (low + high) // 2
```
*   Recursively sort left half:
```
mergeSort(arr, low, mid)
```
* Recursively sort right half:  
```
mergeSort(arr, mid+1, high)
```
* Merge the sorted halves:
```
merge(arr, low, mid, high)
```

Step 3:
```
merge(arr, low, mid, high)
```
1.  Create Temp Array to store merged result.

2.  Set pointers:
```
left = low, right = mid + 1
```
3.  Merge Elements:

*  While
```
left <= mid
```
and
```
right <= high:
```
*   Compare
```
arr[left]
```
and
```
arr[right]
```

4.   Copy Remaining Elements:

*  Copy leftover elements from left half (if any)

*   Copy leftover elements from right half (if any)

5.   Copy from Temp to Original:

*  Copy
```
temp
```
back into
```
arr[low...high]
```
**Recursion continues until full array is sorted.**






Code 👇

````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left_half = arr[:mid]
    right_half = arr[mid:]


    left = merge_sort(left_half)
    right = merge_sort(right_half)

    return merge(left,right)

def merge(left, right):
    temp = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            temp.append(left[i])
            i += 1
        else:
            temp.append(right[j])
            j += 1

    temp.extend(left[i:])
    temp.extend(right[j:])

    return temp


arr = [98, 47, 12, 32, 71, 88]
sorted_arr = merge_sort(arr)
print(sorted_arr)
````

Time Complexity:  
*  **Best Case**: O(n log n), When the array is already sorted or nearly sorted.  
*  **Average Case**: O(n log n), When the array is randomly ordered.  
*  **Worst Case**: O(n log n), When the array is sorted in reverse order.
