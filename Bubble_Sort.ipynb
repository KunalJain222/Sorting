**Bubble Sort**

In Bubble sort:-  
-> we use two loops (one inner and one outer).  
-> the outer loop (say i)is iterated from the last index (n-1) to the first index(0).  
-> the inner loop (say j)is iterated from 0 to i-1.
-> Thus, after each iteration, the last part of the array will become sorted.  
-> After (n-1) iteration, the whole array will be sorted.

```python
def bubble_sort(arr):
  n = len(arr)

  for i in range(n-1,0,-1):
    for j in range(i):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]

  return arr

arr = [37, 49, 91, 58, 12, 6]
sorted_arr = bubble_sort(arr)
print(sorted_arr)
```
